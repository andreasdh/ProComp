# Graphical elements

Here's an overview of extra graphical and structural elements which can be generated through Myst Markdown to use in Jupyterbooks. Three apostrophes (\`\`\`) is used for starting and ending Myst Markdown content. When building structures inside structures, you need to increase the number of apostrophes. For example, if you want a code inside a box inside a tab, the tab has to be initialized with five apostrophes, the box with three apostrophes and the code with three apostrophes.

## Panels

Panels are not generally used, but is a possible way of structuring content if it is seen as useful.

````{panels}
Content of the left panel.

![Jupyter Book Badge](https://img.shields.io/badge/Programming-Computational%20thinking-red)

---

Panel header 2
^^^
Panel body 2
+++
Panel footer 2

````

Panels can be made with the following code. In the first panel we also have a random badge generated through [shields.io](shields.io). 
- The panels are separated with ---
- The first header is marked with ^^^
- The second header is maked with +++

`````
````{panels}
Content of the left panel.

[![Jupyter Book Badge](https://img.shields.io/badge/Programming-Computational%20thinking-red)]

---

Panel header 2
^^^
Panel body 2
+++
Panel footer 2
```
````
`````

We can also specify the number of colomns and the format of the borders:

````{panels}
:column: col-4
:card: border-2
Header A
^^^
Body A
---
Header B
^^^
Body B
---
Header C
^^^
Body C
````

Which is generated with the following code:
`````
````{panels}
:column: col-4
:card: border-2
Header A
^^^
Body A
---
Header B
^^^
Body B
---
Header C
^^^
Body C
````
`````

## Dropdowns
```{admonition} Hint!
:class: tip, dropdown
This is mainly used for extra information and hints.
```

Dropdown menus are created by adding the class _dropdown_ to an admonition (box):

````
```{admonition} Hint!
:class: tip, dropdown
This is mainly used for extra information and hints.
```
````

Code is included in dropdowns like this:
`````
````{admonition} Solution
:class: tip, dropdown
```{code-block} python
for i in range(10):
    print("You are cool!")
```
You can do it like this or you can choose not to.
````
`````

````{admonition} Solution
:class: tip, dropdown
```{code-block} python
for i in range(10):
    print("You are cool!")
```
You can do it like this or you can choose not to.
````

## Tabs
We can create tabs with text and/or coding blocks to separate content, like disciplinary specific such as physics examples, chemistry examples, life science examples and so on.

````{tabbed} C++
Here's how it's done in C++
```{code-block} c++

int main(const int argc, const char **argv) {
  return 0;
}
```
````

````{tabbed} Python
Here's how it's done in Python
```{code-block} python

def main():
    return
```
````

````{tabbed} Java
Here's how it's done in Java
```{code-block} java

class Main {
    public static void main(String[] args) {
    }
}
```
````

````{tabbed} Julia
Here's how it's done in Julia
```{code-block} julia

function main()
end
```
````

````{tabbed} Fortran
Here's how it's done in Fortran
```{code-block} fortran

PROGRAM main
END PROGRAM main
```
````

These tabs are generated with the following code:
`````
````{tabbed} C++
Here's how it's done in C++
```{code-block} c++

int main(const int argc, const char **argv) {
  return 0;
}
```
````

````{tabbed} Python
Here's how it's done in Python
```{code-block} python

def main():
    return
```
````

````{tabbed} Java
Here's how it's done in Java
```{code-block} java

class Main {
    public static void main(String[] args) {
    }
}
```
````

````{tabbed} Julia
Here's how it's done in Julia
```{code-block} julia

function main()
end
```
````

````{tabbed} Fortran
Here's how it's done in Fortran
```{code-block} fortran

PROGRAM main
END PROGRAM main
```
````
`````

    show number on screen
